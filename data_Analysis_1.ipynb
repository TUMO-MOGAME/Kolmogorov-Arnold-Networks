# **3. feature Correlation **

```python
# Calculate correlations for numeric features only
plt.figure(figsize=(12,8))
numeric_df = df.select_dtypes(include=['int64', 'float64'])

# Calculate the correlation matrix
corr_matrix = numeric_df.corr().abs()

# Create a list of pairs with correlation > 0.8
high_corr_list = []
for i in range(len(corr_matrix.columns)):
    for j in range(i):
        if corr_matrix.iloc[i, j] > 0.8:
            colname_i = corr_matrix.columns[i]
            colname_j = corr_matrix.columns[j]
            val = corr_matrix.iloc[i, j]
            high_corr_list.append(f"{colname_i} & {colname_j}: {val:.2f}")

print("Highly Correlated Features:")
for pair in high_corr_list:
    print(pair)
```

When features have a correlation of 1.00, it means they are mathematically identical in your data—they move perfectly in sync.

In a traditional Multilayer Perceptron (MLP), having so many identical inputs can actually lead to "instability" because the model doesn't know which one to "blame" or credit for the output. It's like having ten people all shouting the exact same sentence at you at the same time!

# **I have a lot of features that are highly correlated.**

My goal is to compare:

1. Baseline MLP with all features (likely higher noise/overfitting).

2. Baseline MLP with dropped features (likely cleaner, but still a "black box").

3. KAN Model with all features (to see how it naturally prunes them).

4. KAN Model with dropped features (the most interpretable version).

## ** I need to track two specific metrics:**
1. Accuracy/Loss: Does the model still predict the yield correctly after we remove data?.

2. Interpretability: Can we actually "see" a meaningful relationship in the results, or is it just a mess of numbers?

```python
# This reassigns the column names after stripping the whitespace
df.columns = [col.strip() for col in df.columns]
print("Cleaned Columns:", df.columns.tolist())
```

## **Step 1: Displaying Categorical Features**

```python
categorical_cols = df.select_dtypes(include=['object']).columns

# Display unique values for each to see what we're working with
for col in categorical_cols:
    print(f"{col}: {df[col].unique()[:]}... (Total unique: {df[col].nunique()})")
```

```python
# 1. Create the Full version first
df_full = df.copy()

# 2. Identify Highly Correlated Features (Threshold > 0.9)
corr_matrix = df_full.select_dtypes(include=['number']).corr().abs()
upper_tri = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

# Find columns with correlation greater than 0.90
to_drop = [column for column in upper_tri.columns if any(upper_tri[column] > 0.90)]

# 3. Create the Reduced version automatically
df_reduced = df_full.drop(columns=to_drop)

print(f"Features identified for removal: {to_drop}")
print(f"Full dataset shape: {df_full.shape}")
print(f"Reduced dataset shape: {df_reduced.shape}")
```

# **Redundancy check**

```python
# Select only the object (categorical) columns
obj_cols = df.select_dtypes(include=['object']).columns

# Display the first few rows of these specific columns
print("Categorical Features Preview:")
print(df[obj_cols].head())

# Show how many unique categories are in each
for col in obj_cols:
    print(f"{col}: {df[col].nunique()} unique values")
```

## **Step 2: Creating the Full Encoded Dataset**

```python
# Perform One-Hot Encoding
df_encoded = pd.get_dummies(df, columns=categorical_cols)

# Define our experimental 'Full' dataframe
df_full = df_encoded.copy()

print(f"Final df_full shape: {df_full.shape}")
```

```python
# Perform One-Hot Encoding
df_encoded = pd.get_dummies(df_reduced, columns=categorical_cols)

# Define our experimental 'Full' dataframe
df_reduced = df_encoded.copy()

print(f"Final df_full shape: {df_reduced.shape}")
```
